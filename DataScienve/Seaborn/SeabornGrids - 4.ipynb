```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
grib = sns.PairGrid(iris)
grib.map(plt.scatter)
```

```python
grid = sns.PairGrid(iris)
grid.map_diag(sns.distplot, kde=False)
grid.map_upper(plt.scatter)
grid.map_lower(sns.violinplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips_all = sns.FacetGrid(data=tips, col='time', row='smoker')
```

```python
tips_all = sns.FacetGrid(data=tips, col='time', row='smoker')
tips_all.map(sns.distplot, 'total_bill')
```

```python
tips_all = sns.FacetGrid(data=tips, col='time', row='smoker')
tips_all.map(plt.scatter, 'total_bill', 'tip')
```
