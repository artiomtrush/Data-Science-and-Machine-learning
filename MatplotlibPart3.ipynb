```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
a = np.arange(0,10)
b = a ** 3
```

```python
a
```

```python
b
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, 'r')
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, color = 'red')
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, color = 'red', linewidth = 5)
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, color = 'red', linewidth = 5, alpha = 0.3)
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, color = 'red', linewidth = 1, alpha = 1, linestyle = '--', marker = 'v')
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, color = 'red', linewidth = 1, alpha = 1, linestyle = '--', marker = 'v', markersize = 10)
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, color = 'red', linewidth = 1, alpha = 1, linestyle = '--', marker = 'v', markersize = 10, markerfacecolor = 'black')
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, color = 'red', linewidth = 1, alpha = 1, linestyle = '--', marker = 'v', markersize = 10, 
          markerfacecolor = 'black', markeredgewidth = 3)
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, color = 'red', linewidth = 1, alpha = 1, linestyle = '--', marker = 'v', markersize = 10, 
          markerfacecolor = 'black', markeredgewidth = 1, markeredgecolor = 'blue')
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b, color = 'red', linewidth = 1, alpha = 1, linestyle = '--', marker = 'v', markersize = 10, 
          markerfacecolor = 'black', markeredgewidth = 1, markeredgecolor = 'blue')
axes.set_xlim([0,3])
axes.set_ylim([0,5])
```
