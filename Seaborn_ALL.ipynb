```python
import matplotlib.pyplot as plt
```

```python
grib = sns.PairGrid(iris)
grib.map(plt.scatter)
```

```python
grid = sns.PairGrid(iris)
grid.map_diag(sns.distplot)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
```
