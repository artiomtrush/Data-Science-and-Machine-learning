### 3. Import pandas for csv file reading

### 4. Write the code which allows to show plots in jupiter notebook

### 5. Create a variable and read data from csv file in it

```python
german = pd.read_csv('german_credit_data.csv')
```

### 6. Show first 10 rows from the dataset

```python
german.head(10)
```

### 7. Try to recreate the plot from image below

```python
sns.jointplot(x='Age', y='Credit amount', data=german)
```

### 8. Try to recreate the plot from image below

```python
sns.distplot(german['Duration'], kde=False, color='g')
```

### 9. Try to recreate the plot from image below

```python
sns.set_context('notebook',font_scale = 1)
sns.boxplot(x='Housing', y='Age', data=german)
```

### 10. Try to recreate the plot from image below

```python
sns.swarmplot(x='Housing', y='Age', data=german, palette='plasma')
```

### 11. Try to recreate the plot from image below

```python
sns.countplot(x='Sex', data=german)
```

### 12. Try to recreate the plot from image below

```python
sns.heatmap(german.corr(), cmap='winter')
plt.title('german.corr')
```

### 13. Try to recreate the plot from image below

```python
grid = sns.FacetGrid(data=german, col='Housing')
grid.map(plt.hist, 'Credit amount')
```
