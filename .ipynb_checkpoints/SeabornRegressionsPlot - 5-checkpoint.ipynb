```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',
          markers=['v', 'o'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',
          markers=['v', 'o'], scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, 
           col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, 
           col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, 
           col='day', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, 
           col='day', hue='sex', aspect=0.6, height=4)
```
